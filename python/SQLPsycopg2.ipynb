
Найдите город, пользователи которого в возрасте 36 лет 2021-12-01 числа просмотрели запись на тему covid меньше всего раз относительно других городов. Введите его название на английском.

```python
cursor.execute("""select
  u.city,
  count(post_id)
from
  "user" u
  join feed_action on feed_action.user_id = u.id
  join post on feed_action.post_id = post.id
where
  age = 36
  and date_part('year', feed_action.time) = 2021
  and date_part('day', feed_action.time) = 1
  and date_part('month', feed_action.time) = 12
  and post.topic = 'covid'
group by
  u.city
order by count""")
```

```python
results = cursor.fetchall()
results
```

> Топ постов
Напишите SQL запрос, используя который вы получите:
- топ N постов по лайкам,
- количество пролайкавших пользователей,
- время последнего лайка.

В качестве ответа напишите через пробел первые 3 ID постов из этого топа.

```python
cursor.execute("""SELECT 
    post_id, 
    COUNT(DISTINCT user_id) AS like_count, 
    MAX(time) AS last_like_time
FROM 
    feed_action
WHERE 
    action = 'like'
GROUP BY 
    post_id
ORDER BY 
    like_count DESC
LIMIT 3""")
```

```python
results = cursor.fetchall()
results
```
